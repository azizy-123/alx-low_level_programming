### **Prediction House Price**

OIP.jpeg



- House Price Prediction using Linear Regression
- This project aims to predict house prices based on living space
 using a linear regression model. The dataset used contains
- features such as living space and price. The model is trained
- on a portion of the dataset and evaluated on a test set to
 assess its accuracyitem




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
dataSet = pd.read_csv('kc_house_data.csv')
space = dataSet['sqft_living']
price = dataSet['price']
x = np.array(space).reshape(-1,1)
y = np.array(price)
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
# Fitting linear regression to the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
 # Predictoin the price
 y_pred = regressor.predict(x_test)
```

```python
# Visualizing the training test results
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Visuals for Test Dadaset')
plt.xlabel('Space')
plt.ylabel('Price')
plt.show()
```

```python
# Visualizing the Test results
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Visuals for Test Dataset')
plt.xlabel('Space')
plt.ylabel('Price')
plt.show()
```

```python
# Calculate the Model Accuracy
from sklearn.metrics import mean_squared_error as MSE
mse = MSE(y_test, y_pred)
print('Mean Squared error: ', mse)
```
